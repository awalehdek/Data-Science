Dot Product Calculation
For two vectors a = [a1, a2, ..., an] and b = [b1, b2, ..., bn], their dot product is calculated as: a · b = a1b1 + a2b2 + ... + an*bn.
for example
a = [3, 2, 1]
b = [1, 4, -2]
1- Multiply corresponding elements:
  3 * 1 = 3
  2 * 4 = 8
  1 * (-2) = -2
Sum : 3 + 8 + (-2) = 9
or
a   =  [ 3,  2,  1]
b   =  [ 1,  4, -2]
      ---------------
a[i]*b[i] =  [ 3,  8, -2]
          ---------------
Sum  =  3 + 8 + (-2) = 9


```python
"""
Manual calculation behind dot product:
aa =  | 2   4   6  |    bb =  | 0   1   2  |
      | 1   3   5  |          | 3   4   5  |
      | 10  20  30 |          | 6   7   8  |

Result = | c11  c12  c13 |
         | c21  c22  c23 |
         | c31  c32  c33 |

Where:

c11 = (2*0 + 4*3 + 6*6) = 0 + 12 + 36 = 48
c12 = (2*1 + 4*4 + 6*7) = 2 + 16 + 42 = 60
c13 = (2*2 + 4*5 + 6*8) = 4 + 20 + 48 = 72

c21 = (1*0 + 3*3 + 5*6) = 0 + 9 + 30 = 39
c22 = (1*1 + 3*4 + 5*7) = 1 + 12 + 35 = 48
c23 = (1*2 + 3*5 + 5*8) = 2 + 15 + 40 = 57

c31 = (10*0 + 20*3 + 30*6) = 0 + 60 + 180 = 240
c32 = (10*1 + 20*4 + 30*7) = 10 + 80 + 210 = 300
c33 = (10*2 + 20*5 + 30*8) = 20 + 100 + 240 = 360

Final Result:

| 48   60   72  |
| 39   48   57  |
| 240  300  360 |

"""

np.dot(aa,bb)
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([3, 2])
b = np.array([1, 4])
dot_product = np.dot(a, b)
dot_product
```

```python
plt.figure(figsize=(10, 6))

# Plot vectors
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector a')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector b')

# Plot projections
proj_a = np.dot(a, b) / np.linalg.norm(b) * b / np.linalg.norm(b)
plt.quiver(0, 0, proj_a[0], proj_a[1], angles='xy', scale_units='xy', scale=1, color='g', label='Projection of a on b')

# Add text annotations
plt.text(a[0]/2, a[1]/2, f'a ({a[0]}, {a[1]})', color='r')
plt.text(b[0]/2, b[1]/2, f'b ({b[0]}, {b[1]})', color='b')
plt.text(proj_a[0]/2, proj_a[1]/2, f'Projection', color='g')

# Set plot limits and labels
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'Dot Product Visualization: a · b = {dot_product}')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
