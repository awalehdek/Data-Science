```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from pylab import rcParams
import scipy
from scipy.stats import spearmanr
import warnings
warnings.filterwarnings('ignore')
rcParams['figure.figsize'] = 14,7
plt.style.use('seaborn-whitegrid')
%matplotlib inline
```

### The Spearman Rank Correlation

```python
address = r'Data/mtcars.csv'

cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 
                'wt', 'qsec', 'vs', 'am', 'gear', 'carb', 'y']
```

```python
cars.iloc[:, 0:12].head()
```

```python
sb.pairplot(cars)
```

```python
X = cars[['cyl', 'vs', 'am', 'gear']]
sb.pairplot(X)
```

```python
cyl = cars['cyl']
vs = cars['vs']
am = cars['am']
gear = cars['gear']

spearmanr_coefficient, p_value = spearmanr(cyl, vs)

print('Spearman Rank Correlation Coefficient %0.3f' % (spearmanr_coefficient))
```

```python
spearmanr_coefficient, p_value = spearmanr(cyl, am)

print('Spearman Rank Correlation Coefficient %0.3f' % (spearmanr_coefficient))
```

```python
spearmanr_coefficient, p_value = spearmanr(cyl, gear)

print('Spearman Rank Correlation Coefficient %0.3f' % (spearmanr_coefficient))
```

### Chi-square test for independence

```python
# we assume there is no relantionsip
table = pd.crosstab(cyl, am)
# if the p_value is less < 0.05 we reject the null hypothesis
# p_value 0.013 < 0.05 mean (cyl,am )are related
from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(table.values)
print ('Chi-square statistic %0.3f p_value %0.3f' % (chi2, p))
```

```python
table = pd.crosstab(cyl, vs)

from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(table.values)
print ('Chi-square statistic %0.3f p_value %0.3f' % (chi2, p))
```

```python
table = pd.crosstab(cyl, gear)

from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(table.values)
print ('Chi-square statistic %0.3f p_value %0.3f' % (chi2, p))
```
