```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python


# Address of the CSV file
address = '/content/sample_data/mtcars.csv'

# Read the CSV file
cars = pd.read_csv(address)

# Correct column names
cars.columns = ['car_names', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb', 'extra']

# Display the first few rows
print(cars.head())
```

```python
#  display the summary statistics that decribe a variable's numeric values
# cars.sum() # unsupported error
print(cars.sum(numeric_only=True))
```

```python
# axis=1 means we're operating across columns (horizontally).
#axis=0 (the default) would operate down columns (vertically).


# Select only numeric columns
cars = cars.select_dtypes(include=['int64', 'float64'])

cars.sum(axis=1) # It calculates the sum for each row in the DataFrame.
```

```python
print("# Calculating the sum for each colunm in the DataFrame.")
cars.sum(axis=0)
```

```python
"""
Median:
Calculation: Order the data from least to greatest and find the middle value. If there are two middle values, average them.
Sensitivity: Less sensitive to outliers and skewed data.
Interpretation: Represents the middle value, dividing the data set into two equal halves.
Example: For the data set [1, 2, 3, 4, 100], the median is 3.
"""
# calculates the median value for each numeric column in a DataFrame

cars.median()
```

```python
# Plot Median
# Sample data
data = [1, 2, 3, 4, 100]

# Calculate median
median_value = np.median(data)

# Plotting
plt.figure(figsize=(8, 4))
plt.plot(data, 'bo-', label='Data Points')
plt.axhline(y=median_value, color='r', linestyle='--', label=f'Median = {median_value}')
plt.title('Plot of Data Points with Median')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
"""
Mean (Average):
Calculation: Sum all the values and divide by the number of values.
Sensitivity: Sensitive to outliers and skewed data.
Interpretation: Represents the balance point of the data. If all values were distributed equally, each would be the mean.
Example: For the data set [1, 2, 3, 4, 100], the mean is (1+2+3+4+100)/5 = 22.
"""
# calculating average
cars.mean()
```

```python

# Plot Mean
# Sample data
data = [1, 2, 3, 4, 100]

# Calculate mean
mean_value = np.mean(data)

# Plotting
plt.figure(figsize=(8, 4))
plt.plot(data, 'bo-', label='Data Points')
plt.axhline(y=mean_value, color='g', linestyle='--', label=f'Mean = {mean_value}')
plt.title('Plot of Data Points with Mean')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

# Plot an example of Both **Mean** and **Median**

```python

# Sample data
data = [1, 2, 3, 4, 100]

# Calculate mean and median
mean_value = np.mean(data)
median_value = np.median(data)

# Plotting
plt.figure(figsize=(8, 4))
plt.plot(data, 'bo-', label='Data Points')
plt.axhline(y=mean_value, color='g', linestyle='--', label=f'Mean = {mean_value}')
plt.axhline(y=median_value, color='r', linestyle='--', label=f'Median = {median_value}')
plt.title('Plot of Data Points with Mean and Median')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
mean_value = np.mean(cars)
median_value = np.median(cars)

# Plotting
plt.figure(figsize=(15, 6))
plt.plot(cars, 'bo-', label='Data Points')
plt.axhline(y=mean_value, color='g', linestyle='--', label=f'Mean = {mean_value}')
plt.axhline(y=median_value, color='r', linestyle='--', label=f'Median = {median_value}')
plt.title('Plot of Cars Data Points with Mean and Median')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
"""formula:  = sqrt((x - ) / N)"""
# It calculates the standard deviation for each numeric column in the cars DataFrame
cars.std()
```

```python
std_values = cars.std()

print("MPG Standard Deviation:", std_values['mpg'])
print("Horsepower Standard Deviation:", std_values['hp'])

# Interpret the results
print(f"\nThe MPG varies on average by {std_values['mpg']:.2f} from the mean.")
print(f"The horsepower varies on average by {std_values['hp']:.2f} from the mean.")
```

```python
"""formula:  = (x - ) / (N - 1)"""
# It calculates the variance for each numeric column in the cars DataFrame.
cars.var()
```

```python
var_values = cars.var()
print("MPG Variance:", var_values['mpg'])
print("Horsepower Variance:", var_values['hp'])

# Interpret the results
print(f"\nThe variance in MPG is {var_values['mpg']:.2f} (mpg).")
print(f"The standard deviation in MPG is {np.sqrt(var_values['mpg']):.2f} mpg.")
```
