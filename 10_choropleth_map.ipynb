```python
#import libraries
import pandas as pd
import plotly.express as px
```

```python
#import data
data = pd.read_csv('/content/sample_data/2011_us_ag_exports.csv')
```

```python
#display first few rows
data.head()
```

```python
# create choropleth map for the data
# color will be the column to be color-coded
# locations is the column with spatial coordinates
fig = px.choropleth(data, locations='code',
                    locationmode="USA-states", color='total exports', scope="usa")

fig.show()
```
