# Regressoin Models
## Part 1 - Simple linear regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from pylab import rcParams
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
%matplotlib inline
rcParams['figure.figsize'] = 10,8
```

```python
#np.random.rand(100, 1) generates 100 random numbers between 0 and 1
#Multiplying by 2 stretches that range to 0-2
# Simulating 100 rooms with sizes ranging from 3 to 5units
rooms = 2*np.random.rand(100, 1)+3
rooms[1:10]
```

```python
# Base price is 265, plus 6 times the number of rooms, plus some random noise
# np.random.randn() generates 100 numbers close to zero make the pricing model more realistic
price = 265 + 6*rooms +abs(np.random.randn(100, 1))
price[1:10]
```

```python
plt.plot(rooms, price, 'r^')
plt.xlabel("# of Rooms, 2019 Average")
plt.ylabel("2019 Average Home, 1000s USD")
plt.show()
```

```python
X = rooms
y = price

LinReg = LinearRegression()
LinReg.fit(X,y)
print(LinReg.intercept_, LinReg.coef_)
```

*Simple Algebra*
- y = mx + b
- b = intercept = 265.7

*Estimated Coefficients*
- LinReg.coef_ = [5.99] Estimated coefficients for the terms in the linear regression problem.

### Make a test prediction

```python
y_pred = LinReg.predict(X)
y_pred[:10]
```

```python
# y_pre and y look identical
y[:10]
```

```python
# score is above 95% mean the model performs very well.
print(LinReg.score(X,y))
```

```python
plt.plot(rooms, price, 'r^')
plt.plot(X, y_pred, 'b', label='Simple linear regression')
plt.xlabel("# of Rooms, 2019 Average")
plt.ylabel("2019 Average Home, 1000s USD")
plt.legend(loc="best")
plt.show()
```
